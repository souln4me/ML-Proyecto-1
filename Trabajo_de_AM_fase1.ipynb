```python
"""
El dataset fue cargado utilizando la librería pandas, especificando el separador correspondiente y la codificación adecuada.
Posteriormente, se realizó una inspección inicial para conocer la estructura de los datos, incluyendo el número de registros y
variables disponibles, así como una visualización preliminar de las primeras observaciones.
"""

df = pd.read_csv("dataset_elpino.csv", sep=";", encoding="utf-8", engine="python")

print("Dimensiones del dataset:", df.shape)
print("\nTipos de datos:\n", df.dtypes)

df.head()
```

```python
"""
Se realizó una limpieza de las variables de diagnósticos y procedimientos, extrayendo únicamente el código asociado a
cada registro. Esto permitió eliminar descripciones textuales redundantes y estandarizar la información, facilitando su
posterior procesamiento por modelos de aprendizaje automático.
"""

for col in df.filter(like="Diag").columns:
    df[col] = df[col].apply(extraer_codigo)

for col in df.filter(like="Proced").columns:
    df[col] = df[col].apply(extraer_codigo)

df.head()
```

```python
"""
Debido a la alta dispersión del dataset original, se consolidaron múltiples columnas de diagnósticos y procedimientos
en representaciones compactas por paciente. Para ello, se concatenaron los códigos presentes en cada fila, excluyendo
valores nulos, generando nuevas variables que representan el conjunto de eventos clínicos asociados a cada individuo.
"""


# columnas
diag_cols = df.filter(like="Diag").columns
proc_cols = df.filter(like="Proced").columns

# función
def unir_codigos(fila, columnas):
    return " ".join([str(x) for x in fila[columnas] if pd.notnull(x)])

# nuevas variables
df["Diagnosticos"] = df.apply(lambda row: unir_codigos(row, diag_cols), axis=1)
df["Procedimientos"] = df.apply(lambda row: unir_codigos(row, proc_cols), axis=1)

# verificación
df[["Diagnosticos", "Procedimientos"]].head()
```

```python
"""
Se eliminaron espacios en blanco residuales en las columnas consolidadas de diagnósticos y procedimientos para asegurar
consistencia en la representación textual de los códigos.
"""

df["Diagnosticos"] = df["Diagnosticos"].str.strip()
df["Procedimientos"] = df["Procedimientos"].str.strip()
```

```python
"""
Se realizó una validación descriptiva de la nueva representación por paciente, estimando la cantidad de diagnósticos
asociados a cada registro. Esto permitió confirmar que la transformación preservó la información clínica relevante.
"""

df["Diagnosticos"].apply(lambda x: len(x.split())).describe()
```

```python
"""
Se utilizó CountVectorizer en modo binario para representar la presencia o ausencia de códigos clínicos en cada paciente,
evitando sesgos asociados a la frecuencia de aparición de los mismos.
"""


from sklearn.feature_extraction.text import CountVectorizer

# Inicializar vectorizador en modo binario (presencia/ausencia de códigos)
vectorizer = CountVectorizer(binary=True)

# Ajustar y transformar los diagnósticos en una matriz numérica
X = vectorizer.fit_transform(df["Diagnosticos"])

# Mostrar dimensiones de la matriz resultante
print("Shape matriz (pacientes, códigos):", X.shape)
```

```python
"""
Se realizó un preprocesamiento de la variable objetivo, extrayendo únicamente el código del GRD y eliminando
la descripción asociada. Esto permitió trabajar con etiquetas más compactas y consistentes para el entrenamiento
de los modelos de clasificación.
"""


y = df["GRD"].apply(lambda x: str(x).split(" - ")[0])
```
